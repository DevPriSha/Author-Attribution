```python
new_pcf_gen_common = pcf_gen_common.drop(["Date_x", "Author_x","Content_x"],axis = 1)
```

```python
new_pcf_gen_common
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
docs = new_pcf_gen_common["Content"].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(docs)
```

```python
k = 10
model = KMeans(n_clusters = k, init = 'k-means++', max_iter= 100,n_init= 1)
model.fit(features)
```

```python
new_pcf_gen_common['cluster'] = model.labels_
```

```python
new_pcf_gen_common
```

```python
clusters = new_pcf_gen_common.groupby('cluster')

for cluster in clusters.groups:
    f = open('cluster'+str(cluster)+ '.csv', 'w', encoding='utf-8') # create csv file
    data_new = clusters.get_group(cluster)[['Author','Content']] # get title and overview columns
    f.write(data_new.to_csv()) # set index to id
    f.close()
```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]:
        print (' %s' % terms[j])
    print('------------')
```
