```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
docs = new_pcf_gen_common["Content"].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(docs)
```

```python
features.shape
```

```python
from sklearn.cluster import KMeans
```

```python
k = 10
model = KMeans(n_clusters = k, init = 'k-means++', max_iter= 100,n_init= 1)
model.fit(features)
```

```python
pred_labels = model.labels_
```

```python
from sklearn import metrics
```

```python
ss = metrics.silhouette_score(features.toarray(),pred_labels, metric = "euclidean",sample_size = 1000, random_state = 200)
```

```python
print(ss)
```

```python
dbi = metrics.davies_bouldin_score(features.toarray(),pred_labels)
```

```python
print(dbi)
```

## Implementing dbscan with elbow method

```python
k = min_samples if min_samples > 2 else 2
nbrs = NearestNeighbors(n_neighbors=k).fit(features)
distances, indices = nbrs.kneighbors(features)
```

```python
distances.shape
```

```python
for enum, row in enumerate(distances[:]):
    print('observation' + str(enum)+ ':' + str([round(x,2) for x in row]))
```

```python
new_pcf_gen_common["knn_farthest_dist"] = distances[:,-1]
```

```python
new_pcf_gen_common
```

```python
new_pcf_gen_common.sort_values("knn_farthest_dist", ascending = False).reset_index()[["knn_farthest_dist"]].plot()
plt.xlabel("index")
plt.ylabel("distance")
plt.grid(True)
plt.show()
```

```python
km = DBSCAN(eps = 0.9, min_samples = min_samples)
cluster_dbscan = km.fit(features)
cluster = cluster_dbscan.labels_
```
