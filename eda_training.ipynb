```python
# import stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
pcf_gen = pd.read_csv("./Data/pcf-general.csv")
spl_gen_1 = pd.read_csv("./Data/spl-gen-1.csv")
spl_gen_2 = pd.read_csv("./Data/spl-gen-2.csv")
```

```python
# appending spl_gen_2 to spl_gen_1
spl_gen = pd.concat([spl_gen_1, spl_gen_2], axis=0, ignore_index=True)
```

## 1. Exploratory Data Analyis

```python
spl_gen
```

```python
pcf_gen
```

```python
# list of all authorID common to both dataframes
authorID = list(set(pcf_gen['AuthorID']).intersection(set(spl_gen['AuthorID'])))
print(authorID)
print("No. of users common to both servers: ", len(authorID))
```

```python
#create a dataframe with AuthorID and the number of content in each server
df = pd.DataFrame(columns=['AuthorID', 'Author', 'pcf_gen', 'spl_gen'])
df['AuthorID'] = authorID
#find author name for each authorID
df['Author'] = df['AuthorID'].apply(lambda x: pcf_gen[pcf_gen['AuthorID'] == x]['Author'].values[0])
df['pcf_gen'] = df['AuthorID'].apply(lambda x: pcf_gen[pcf_gen['AuthorID'] == x].shape[0])
df['spl_gen'] = df['AuthorID'].apply(lambda x: spl_gen[spl_gen['AuthorID'] == x].shape[0])
#sort df by sum of content in both servers
df = df.sort_values(by=['pcf_gen', 'spl_gen'], ascending=False)
df
```

```python
#save df to csv
df.to_csv("./Data/common-users.csv", index=False)
```

```python
#plot the number of content in each server
plt.figure(figsize=(10, 5))
plt.bar(df['Author'], df['pcf_gen'], label='pcf_gen')
plt.bar(df['Author'], df['spl_gen'], bottom=df['pcf_gen'], label='spl_gen')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# plot the number of content in pcf_gen by common users
plt.figure(figsize=(10, 5))
plt.bar(df['Author'], df['pcf_gen'], label='pcf_gen')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# plot the number of content in spl_gen by common users
plt.figure(figsize=(10, 5))
plt.bar(df['Author'], df['spl_gen'], label='spl_gen')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
